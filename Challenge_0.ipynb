```python
import numpy as np
import pandas as pd
```

```python
df_0 = pd.read_csv('workshops.csv')
df_0
```

```python
df_0['Year'].fillna(method='ffill', inplace=True)
df_0['Month'].fillna(method='ffill', inplace=True)
df_0
```

```python
df_1 = df_0[pd.notnull(df_0['Earnings'])].copy()
df_1
```

```python
def as_date(row, col):
    year = int(row['Year'])
    month = row['Month']
    day = int(row[col])
    ts = f'{month} {day}, {year}'
    return pd.to_datetime(ts, format='%B %d, %Y')

df_1['start'] = df_1.apply(as_date, axis=1, args=('Start',))
df_1['end'] = df_1.apply(as_date, axis=1, args=('End',))
df_1
```

```python
def topic(name):
    if 'go' in name:
        return 'go'
    if 'python' in name:
        return 'python'

df_1['topic'] = df_1['Name'].str.lower().apply(topic)
df_1
```

```python
df_1['earnings'] = pd.to_numeric(
    df_1['Earnings'].str.replace(r'[$,]', '')
).astype(np.float64)
df_1
```

```python
df_1 = df_1[['start', 'end', 'Name', 'topic', 'earnings']]
df_1.rename(columns={'Name': 'name'}, inplace=True)
df_1
```
